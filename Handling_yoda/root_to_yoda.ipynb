```python
import ROOT
from array import array

# Define histogram properties
histogram_properties = {
    'mjj': {'title': 'mjj', 'bins': 3, 'bin_edges': array('d', [500, 800, 1200, 1600])},
    'pt': {'title': 'PT', 'bins': 3, 'bin_edges': array('d', [200, 250, 350, 700])},
    'delta_phi_jj': {'title': 'deltaphi', 'bins': 4, 'bin_edges': array('d', [-3.14, -2, 0, 2, 3.14])}
}

# Define bins and years
bins = ['bin1', 'bin2', 'bin3']
years = ['2016', '2016pre', '2017', '2018']

# Loop through each year
for year in years:
    # Create a new ROOT file for the current year
    output_file = ROOT.TFile(f"TotalSig_{year}.root", "RECREATE")

    # Loop through each histogram
    for hist_name, properties in histogram_properties.items():
        # Create a new histogram
        hist = ROOT.TH1D(hist_name, properties['title'], properties['bins'], properties['bin_edges'])
        for l in range(len(bins)):
            b = bins[l]
            print(year,b)
            integral = 0
            # Assuming file is defined somewhere
            for i in range(len(file.GetListOfKeys())):
                Folder = file.GetListOfKeys()[i].ReadObj()
                print(f"first:{Folder}")
                if f'SR_{year}_SR_{b}_{year}_prefit' in Folder.GetName():
                    print(f"then:{Folder}")
                    obj = Folder.Get('TotalSig')
                    integral = obj.Integral()

            # Set the content of the central bin to the integral value
            hist.SetBinContent(l+1, integral)
            
        print(l,i)
        # Write the histogram to the output file
        print(hist.Integral())
        hist.Scale(1/hist.Integral())
        hist.Write()
        print(hist.Integral())
    # Close the output file for the current year
    output_file.Close()
```

```python
import uproot
import yoda

def root_to_yoda(input_root_file, output_yoda_file):
    # Open the ROOT file and access the desired histogram                                                                                                                                                          
    with uproot.open(input_root_file) as f:
        hist = f["pt"]

        # Extract the bin content and errors                                                                                                                                                                       
        bin_centers = hist.axis().centers()
        bin_contents = hist.values()
        bin_errors = hist.errors()
        bin_edges = hist.axis().edges()
        print("errors: ",bin_errors)
        print("contents: ",bin_contents)
        print("centers: ",hist.axis().centers())
        print("edges: ",bin_edges)
        scatter = yoda.Scatter2D()
        for i in range(len(bin_contents)):
            x_min = bin_edges[i]
            x_max = bin_edges[i + 1]
            center = (x_min + x_max) / 2
            content = bin_contents[i]
            error = bin_errors[i]
            # Add point with x and y errors
            scatter.addPoint(yoda.Point2D(center, content, xerrs=(center - x_min, x_max - center), yerrs=error))
            print(scatter.yErrs())

    # Write the Scatter2D object to a YODA file                                                                                                                                                                    
    with open(output_yoda_file, "w") as f:
        yoda.writeYODA(scatter, f)

# Example usage:                                                                                                                                                                                                   
input_files = ["TotalSig_2016pre"]#,"TotalSig_2016","TotalSig_2016pre", "TotalSig_2017", "TotalSig_2018"]
for inputfile in input_files:
    input_root_file = f'{inputfile}.root'
    output_yoda_file = f"{inputfile}.yoda"
    root_to_yoda(input_root_file, output_yoda_file)
    print(f'{output_yoda_file} has been  created')
```

```python
import yoda

def merge_yoda_files(input_yoda_files, output_yoda_file):
    # Create an empty list to hold YODA objects
    yoda_objects = []

    # Loop through each input YODA file
    for input_file in input_yoda_files:
        # Load the YODA file
        data = yoda.read(input_file)

        # Append the YODA data to the list
        yoda_objects.append(data)

    # Merge the YODA objects
    merged_data = yoda.core.merge(*yoda_objects)

    # Write the merged data to the output YODA file
    yoda.write(output_yoda_file, merged_data)

# Example usage:
input_yoda_files = ["outdelta_phi_jj.yoda", "outpt.yoda", "outmjj.yoda"]
output_yoda_file = "merged_output.yoda"
merge_yoda_files(input_yoda_files, output_yoda_file)


```

```python
def merge_yoda_files(input_yoda_files, output_yoda_file):
    # Open the output YODA file for writing
    with open(output_yoda_file, "w") as output_file:
        # Loop through each input YODA file
        for input_file in input_yoda_files:
            # Open the input YODA file for reading
            with open(input_file, "r") as input_file:
                # Read the contents of the input YODA file
                yoda_contents = input_file.read()

                # Write the contents to the output YODA file
                output_file.write(yoda_contents)

# Example usage:
input_yoda_files = ["outdelta_phi_jj.yoda", "outpt.yoda", "outmjj.yoda"]
output_yoda_file = "merged_output.yoda"
merge_yoda_files(input_yoda_files, output_yoda_file)
```

```python
# Create a ROOT canvas
canvas = ROOT.TCanvas("canvas", "Rivet Analysis Histograms", 800, 600)
yoda_file = "test.yoda"
data = yoda.readYODA(yoda_file)
# Define a function to convert YODA histograms to ROOT histograms
def yoda_to_root(yoda_hist):
    # Create a ROOT histogram with the same binning as the YODA histogram
    nbins = len(yoda_hist.bins())
    x_min = yoda_hist.xMin()
    x_max = yoda_hist.xMax()
    root_hist = ROOT.TH1F(f'{yoda_hist.name()}', f'{yoda_hist.name()}', nbins, x_min, x_max)
    
    # Fill the ROOT histogram with the YODA histogram data
    for i, b in enumerate(yoda_hist.bins()):
        print(b)
        root_hist.SetBinContent(i+1, b.height())
        root_hist.SetBinError(i+1, b.heightErr())
    
    return root_hist

# Initialize a ROOT histogram for the sum
summed_hist = None

# Loop over the YODA data, find and sum "photon pt" histograms
for path, hist in data.items():
    if "pt" in path:
        root_hist = yoda_to_root(hist)
        if summed_hist is None:
            summed_hist = root_hist.Clone("summed mjj")
            summed_hist.SetTitle("Summed mjj")
        else:
            summed_hist.Add(root_hist)

# Draw the summed histogram
if summed_hist:
    summed_hist.Draw()
    canvas.SaveAs("summed_delta_phi.png")
else:
    print("No 'photon pt' histograms found in the YODA file.")
```
